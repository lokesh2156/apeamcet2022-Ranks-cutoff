#### Table Creation

```python
try :
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        
        cursor.execute("DROP TABLE IF EXISTS ranks_cutoff_transformed")
        print("Creating table,...")
        cursor.execute("CREATE TABLE ranks_cutoff_transformed (S_No int,College_Code varchar(10),College_Name TEXT,College_Type varchar(5),Region varchar(5),District varchar(15),Place TEXT,COED varchar(10),Affliate_Univ varchar(15),ESTD int,Branch varchar(10),Gender varchar(7),Category varchar(10),`Rank` int,Fee int)")
        print('Table Name:'+'ranks_cutoff_transformed')
        print('Table Created !')
        for i,row in tqdm(df.iterrows()):
            sql = "INSERT INTO APEAMCET2022.ranks_cutoff_transformed(S_No,College_Code,College_Name,College_Type,Region,District,Place,COED,Affliate_Univ,ESTD,Branch,Gender,Category,`Rank`,Fee) VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            if i%10000 == 0:
                print(str(i)+" Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except mysql.Error as e:
    print("Error while connecting to MySQL", e)

finally:
    # Close the cursor and connection
    cursor.close()
    conn.close()
```
