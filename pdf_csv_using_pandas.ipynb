# PDF To CSV By Using Pandas

```python
# Importing Libriaries
import pdfplumber
import pandas as pd
```

```python
# Reading pdf
pdf = pdfplumber.open(r"D:\Data sets\APEAMCET-2022\APEAPCET_2022_Cutoff.pdf")
```

```python
# pdf pages size 
size = len(pdf.pages)
size
```

```python
# All Records in pdf have same header so I'm using 
# 1st page data to make header
page0 = pdf.pages[0]
# Extracting Tables from page
table = page0.extract_table()
len(table)
```

```python
# Checking Data weather correct not by looking some records I used 3 rows
print(table[:3])
```

```python
# Here Table is nothing but Multi Dimensional Array 
# So use Indexes for fetching columns Everypage has header so we will add that header to data frame only once
df = pd.DataFrame(columns = table[0])
```

```python
# Main dataframe i.e, header  
df
```

```python
# Each page data expect header this is each_page_data
# We are specifying colums are headers (extracted in above cell)
each_page_data = pd.DataFrame(table[1:], columns=df.columns)
each_page_data
```

```python
# 1st Page Statistics
each_page_data.describe()
```

```python
# Data types
df.dtypes
```


## Main task
- **We need Each Record in Each Table of Each Page** 
- **So Do Traversing through each page,each table,each record**
- **We already have header in df so don't need header here**
### Note: 
    - **We need to append each page to data main df so I used concat method (append is depriciated)**
    - **we need to give exisiting df then new df which is going to append to exisiting df**

```python
# Traverse throgh each page
for i in range(size):
    # storing each page data 
    page = pdf.pages[i]
    # Extracting tabular form data from page 
    table = page.extract_table()
    # storing as dataframe from every table from 1 index (header is already ther)
    each_page_data = pd.DataFrame(table[1:], columns=df.columns)     
    # Adding each df to main df by using concat (Eg :sum of numbers in array)
    df = pd.concat([df, each_page_data], ignore_index=True)
```
