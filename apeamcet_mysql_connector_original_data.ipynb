#### Table Creation

```python
try:
    # db credentials & connection
    conn = msql.connect(host='localhost', database='APEAMCET2022', user='root', password='Lokesh&&5655')
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        
        # Retrieving Records from csv file
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        
        # Table Creation
        cursor.execute('DROP TABLE IF EXISTS rankscutoff;')
        print('Creating table....')
        cursor.execute("CREATE TABLE rankscutoff(S_No int,College_Code varchar(10), College_Name TEXT,College_Type varchar(5),Region varchar(5),District varchar(5),Place TEXT,COED varchar(10),Affliate_Univ varchar(15),ESTD int,Branch varchar(10),OC_BOYS int,OC_GIRLS int,SC_BOYS int,SC_GIRLS int,ST_BOYS int,ST_GIRLS int,BC_A_BOYS int,BC_A_GIRLS int,BC_B_BOYS int,BC_B_GIRLS int,BC_C_BOYS int,BC_C_GIRLS int,BC_D_BOYS int,BC_D_GIRLS int,BC_E_BOYS int,BC_E_GIRLS int,OC_EWS_BOYS int,OC_EWS_GIRLS int,FEE int)") # SQL Command
        print('Table Name:'+'rankscutoff')
        print('Table Created !')
        
except msql.Error as e:
            print("Error while connecting to MySQL", e)
```

#### Inserting Records

```python
try:
    # Traversing through each record i = index, row = data (tuple) 
    for i, row in tqdm(df.iterrows(), total=len(df)):
        # string as placeholders for all diff data types  
        sql = "INSERT INTO rankscutoff (S_No, College_Code, College_Name, College_Type, Region, District, Place, COED, Affliate_Univ, ESTD, Branch, OC_BOYS, OC_GIRLS, SC_BOYS, SC_GIRLS, ST_BOYS, ST_GIRLS, BC_A_BOYS, BC_A_GIRLS, BC_B_BOYS, BC_B_GIRLS, BC_C_BOYS, BC_C_GIRLS, BC_D_BOYS, BC_D_GIRLS, BC_E_BOYS, BC_E_GIRLS, OC_EWS_BOYS, OC_EWS_GIRLS, FEE) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"
        
        # Check if the number of placeholders and the number of elements in the row match
        if len(row) != sql.count("%s"):
            raise ValueError(f"Number of placeholders in SQL query ({sql.count('%s')}) does not match the number of elements in the row ({len(row)})")

        # Inserting each record as a tuple
        cursor.execute(sql, tuple(row))

        # Records count (optional)
        if i % 100 == 0:
            print(f"{i} Record inserted")

    # Commit all changes after the loop
    conn.commit()

except msql.Error as e:
    print("Error while connecting to MySQL", e)

finally:
    # Close the cursor and connection
    cursor.close()
    conn.close()
```
