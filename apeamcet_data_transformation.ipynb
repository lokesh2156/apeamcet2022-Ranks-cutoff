```python
import pandas as pd


# Get the columns representing category and gender ranks
category_columns = df.columns[11:-1]  # Exclude the last column 'FEE'

# Melt the DataFrame to combine ranks into a single column
melted_df = pd.melt(df, id_vars=['S No', 'College_Code', 'College_Name', 'College_Type', 'Region', 'District', 'Place', 'COED', 'Affliate_Univ', 'ESTD', 'Branch', 'FEE'], value_vars=category_columns, var_name='Category_Gender', value_name='Rank')

# Extract Category and Gender from 'Category_Gender' column
# Eg : OC_BOYS
melted_df['Category'] = melted_df['Category_Gender'].str[:-5] # OC upto _
melted_df['Gender'] = melted_df['Category_Gender'].str[-4:]   # BOYS

# Remove rows where the rank is NaN or 0 (assuming 0 means data is missing)
melted_df = melted_df.dropna(subset=['Rank'])
melted_df = melted_df[melted_df['Rank'] != 0]

# Drop the 'Category_Gender' column as it's no longer needed
# We have separate columns now Category and Gender
melted_df.drop(columns=['Category_Gender'], inplace=True)

final_output = melted_df.sort_values(by=['S No', 'Gender', 'Category']).reset_index(drop=True)

# Merge with original data to get the 'FEE' column back
# This is like LEFT JOIN with melted_df with original df based on S_No, College_Code, Branch  
final_output = pd.merge(final_output, df[['S No', 'College_Code', 'Branch', 'FEE']], on=['S No', 'College_Code', 'Branch'], how='left')

print(final_output)
```
